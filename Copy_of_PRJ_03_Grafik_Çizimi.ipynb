# Görev 1

```python
# veri yükle, verinin tüm satırlarını göster (24 satır)
url = "https://docs.google.com/spreadsheets/d/e/2PACX-1vS15GgqoaSco85tiyALex0yRXXUM6EzY5vtuN5MbZJQxbqlhNV89tMGUCsLbXCGespULDUJzjitKNAC/pub?gid=1377583252&single=true&output=csv"

import pandas as pd
#df = pd.read_csv("veri-satisi.csv", encoding="utf-8")
df = pd.read_csv(url)
print(df)
```

# Görev 2

```python
# zigana dağcılık firmasının gelirlerini çizgi grafiği ile çiz
import matplotlib.pyplot as plt
zigana_data = df[df["Firma"] == "Zigana Dağcılık"]
plt.figure(figsize=(10, 6))

plt.plot(zigana_data["Ay"], zigana_data["Gelir"], 'o-', label="Gelir")
plt.show()
```

# Görev 3

```python
# Beypınarı holding firmasının gelir ve giderini aynı grafikte çizgi grafiği olarak göster, gelir yeşil renkli, gider kırmızı renkli olsun

beypinari_data = df[df["Firma"] == "Beypınarı Holding"]
plt.figure(figsize=(10, 6))
plt.plot(beypinari_data["Ay"], beypinari_data["Gelir"], 'o-', color='green', label="Gelir (Yeşil)")
plt.plot(beypinari_data["Ay"], beypinari_data["Gider"], 'o-', color='red', label="Gider (Kırmızı)")

plt.title("Beypınarı Holding Gelir ve Giderleri", fontsize=14)
plt.xlabel("Ay", fontsize=12)
plt.ylabel("Tutar (₺)", fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle="--", alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```

# Görev 4

```python
# dataframe tablosuna kar adında yeni bir kolon ekle


df["Kar"] = df["Gelir"] - df["Gider"]


print(df.head())
```

# Görev 5

```python
# yanyana 2 pasta grafiği çiz, ilk pasta grafiğinde Beypınarı holding firmasının çeyrek dönemlere göre gelirini, 2. pasta grafiğinde çeyrek dönemlere göre giderini gösteren pasta grafiğini çiz
#beypinari_data = df[df["Firma"] == "Beypınarı Holding"]

gelir_by_ceyrek = beypinari_data.groupby("Çeyrek")["Gelir"].sum()
gider_by_ceyrek = beypinari_data.groupby("Çeyrek")["Gider"].sum()

fig, axes = plt.subplots(1, 2, figsize=(14, 7))

axes[0].pie(
    gelir_by_ceyrek,
    labels=gelir_by_ceyrek.index,
    autopct='%1.1f%%',
    colors=['#4caf50', '#8bc34a', '#cddc39', '#ffeb3b']
)
axes[0].set_title("Beypınarı holding - Çeyrek Dönemlere Göre Gelir")

axes[1].pie(
    gider_by_ceyrek,
    labels=gider_by_ceyrek.index,
    autopct='%1.1f%%',
    colors=['#f44336', '#e57373', '#ffcdd2', '#ff8a80']
)
axes[1].set_title("Beypınarı holding - Çeyrek Dönemlere Göre Gider")

plt.tight_layout()
plt.show()
```

# Görev 6

```python
# Zigana firmasının gelir ve gider değerlerinin scatter grafiğini çiz

zigana_data = df[df["Firma"] == "Zigana Dağcılık"]

plt.figure(figsize=(8, 6))
plt.scatter(zigana_data["Gelir"], zigana_data["Gider"], color='blue', label="Gelir vs Gider")

plt.title("Zigana Dağcılık Gelir ve Gider Dağılımı", fontsize=14)
plt.xlabel("Gelir (₺)", fontsize=12)
plt.ylabel("Gider (₺)", fontsize=12)

plt.show()
```

# Görev 7

```python
# beypınarı firmasının giderlerini sütun grafiği ile çiz

plt.figure(figsize=(10, 6))
plt.bar(beypinari_data["Ay"], beypinari_data["Gider"], color='red', label="Gider")

plt.title("Beypınarı Holding Giderleri", fontsize=14)
plt.xlabel("Ay", fontsize=12)
plt.ylabel("Gider (₺)", fontsize=12)

plt.show()
```

# Görev 8

```python
# beypınarı firmasının gelir-giderlerinin sütun grafiğini karşılaştırmalı olarak çiz
import numpy as np

x = np.arange(len(beypinari_data["Ay"]))
width = 0.4

plt.figure(figsize=(12, 6))
plt.bar(x - width / 2, beypinari_data["Gelir"], width, color='green', label="Gelir")
plt.bar(x + width / 2, beypinari_data["Gider"], width, color='red', label="Gider")
plt.title("Beypınarı Holding Gelir ve Gider Karşılaştırması", fontsize=14)
plt.xlabel("Ay", fontsize=12)
plt.ylabel("Tutar (₺)", fontsize=12)
plt.xticks(x, beypinari_data["Ay"], rotation=45)
plt.legend()

plt.tight_layout()
plt.show()
```

# Görev 9

```python
# Firmaların çeyrek dönemler bazında kar durumlarını sütun grafiği olarak çiz

kar_by_ceyrek = df.groupby("Çeyrek")["Kar"].sum()

plt.figure(figsize=(10, 6))
plt.bar(kar_by_ceyrek.index, kar_by_ceyrek.values, color='blue', label="Kar")

plt.title("Firmaların Çeyrek Dönemler Bazında Kar Durumu", fontsize=14)
plt.xlabel("Çeyrek", fontsize=12)
plt.ylabel("Toplam Kar (₺)", fontsize=12)
plt.xticks(rotation=0)
plt.legend()

plt.tight_layout()
plt.show()
```
